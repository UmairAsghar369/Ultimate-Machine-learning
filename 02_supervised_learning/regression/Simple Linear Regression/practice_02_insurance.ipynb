```python
#Importing libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
#Reading dataset
df=pd.read_csv('../../../datasets/Insuranse dataset(SLR).csv')
```

```python
#Showing how the variables are scattered
plt.figure(figsize=(12,6))
sns.pairplot(df,x_vars=['Age'],y_vars=['Premium'],size=7,kind='scatter')
plt.xlabel('Years')
plt.ylabel('Salary')
```

```python
sns.regplot(x="Age", y="Premium", data=df)
```

```python
X = df[["Age"]]
y = df["Premium"]
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y, train_size=0.7,random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
lr.fit(X_train,y_train)
```

```python
y_predict = lr.predict(X_test)
y_predict
```
