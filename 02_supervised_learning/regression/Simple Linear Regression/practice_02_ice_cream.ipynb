```python
# simple linear regression ke lie dono variables ka numeric hona zarori ha 
# so we will convert object (DATE) into numeric
```

```python
df['DATE'] = pd.to_datetime(df['DATE'])        # convert to datetime
df['DATE_num'] = df['DATE'].map(lambda x: x.toordinal())  # convert to number
```

```python

#Showing how the variables are scattered
plt.figure(figsize=(12,6))
sns.pairplot(df,x_vars=['DATE'],y_vars=['Sales'],size=7,kind='scatter')
plt.xlabel('Dates')
plt.ylabel('Sales')
```
