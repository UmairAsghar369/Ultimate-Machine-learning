```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python

df=pd.read_csv('../../../datasets/house_price_prediction(MLR).csv')
```

```python
df.dropna(inplace=True)
```

```python
df.describe()
```

```python
df['date'] = pd.to_datetime(df['date'])
df['year'] = df['date'].dt.year
df['month'] = df['date'].dt.month
df['day'] = df['date'].dt.day
df.drop('date', axis=1, inplace=True)
```

```python
df.country.value_counts()
```

```python
df[['street','city','statezip','country']].nunique()
```

```python
df = df.drop(['street', 'country'], axis=1)  # street almost unique, country is the same for all rows
df = df.drop(['statezip'], axis=1)  # statezip or city is enough, we can drop one of them, atlest for this simple model
```

```python
pd.get_dummies(df['city'], drop_first=True)
```

```python
df = df.join(pd.get_dummies(df['city'], dtype=int, drop_first=True)).drop('city', axis=1)
```
